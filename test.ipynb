```python
import cv2


# Open the video file
video_path = "video_2024-03-04_22-46-54.mp4"
cap = cv2.VideoCapture(video_path)

# Get video properties
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Create VideoWriter object to save the output
output_path = 'res_for_plate'+video_path  # Change the output file path and format if needed
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # You can change the codec according to your system
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

# Loop through the video frames
while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()

    if success:
        # Run YOLOv8 inference on the frame
        results = model.predict(frame, save=True)

        # Visualize the results on the frame
        annotated_frame = results[0].plot()

        # Save the annotated frame to the output video file
        out.write(annotated_frame)

        # Display the annotated frame
        cv2.imshow("YOLOv8 Inference", annotated_frame)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # Break the loop if the end of the video is reached
        break

# Release the video capture object, VideoWriter, and close the display window
cap.release()
out.release()
cv2.destroyAllWindows()
```
